## Отчет по тестовому заданию

### Шаги/идеи
Изначально я представлял решение задачи, как какое-то проставление весов для каждого слова и определение тональности по наличию (плохих|хороших) слов, но где найти такой китайский словарь.

После я прочитал некоторое количество статей про нейронные сети, и пришло в голову интересное (или не очень) решение:
- cделать синтаксический разбор предложения. Правда перед этим нужно будет перевести текст (https://pypi.org/project/py-translator/) на (английский|русский) язык, т.к. я не нашел инструмент для синт. разбора китайского языка (вот для англ. рус. https://api.ispras.ru/). 
- после нужно убрать из него высказывания, не имеющие смысла, либо проставить каждому высказыванию веса, определяющие значимость конкретного отношения, пользуясь корпусом синт. разбора (нашел только для русского http://www.ruscorpora.ru/instruction-syntax.html#%D0%A1%D0%B8%D0%BD%D1%82%D0%B0%D0%BA%D1%81%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%B0%D1%8F)
<img src="image.png">
Например: здесь, Yes (вводное синт. отн.) будет иметь меньший вес чем it tastes good (предикативное)) и т.о. не будет влиять на тональность ревью.
- и в NN пойдут на вход уже конкретные высказывания (с весами, если есть), вместо целого текста ревью. 
Также можно заставить NN запоминать конкретные последовательности синтаксических частей. Т.е. будет иметь большую значимость описание чего либо, либо утверждение: [подлежащее, сказуемое], [подлежащее, определение], ... Но я не знаю, как это можно сделать :_(

Но все это не работает, т.к. сначало нужно перевести все ревью на английский, и сделать синтаксический разбор. Это очень затратные и долгие операции: 500 символов переводится в среднем 1 секунду, синт. разбор также длится долго, притом все это нужно будет сделать во время предсказания, что точно не подходит. 

После того, как я понял, что эта идея провальная, я решил почитать про использование CNN для sentiment analisys:

https://habr.com/en/post/249215/

https://habr.com/en/company/mailru/blog/417767/

https://habr.com/en/company/ods/blog/353060/

Там писали про точность 60-70%, но я почему-то чувствовал, что на китайском языке будет еще хуже. А как сказал преподаватель, я тебе и сам могу предсказать с точностью 60%.

Потом прочитал про использование RNN:

https://habr.com/en/company/dca/blog/274027/

https://towardsdatascience.com/a-beginners-guide-on-sentiment-analysis-with-rnn-9e100627c02e

https://www.liip.ch/en/blog/sentiment-detection-with-keras-word-embeddings-and-lstm-deep-learning-networks

Решил остановиться на решении с RNN. 

### Анализ данных
Не проводил очистку данных от мусора (бесполезных слов), т.к. не понятно, что в китайском языке есть мусор, и предположил, что RNN будет так только лучше понимать контекст. Поэтому я только очистил все не буквенные символы.

Для обучения модели использовал только ревью с оценкой 1 и 5, т.к при обучении сети на всех оценках, accuracy держалось уверенно в районе 50%. 


### Описание модели
Модель RNN состоит из двух слоев LSTM и логистической регрессии (Dense(1), Activation('sigmoid')).
